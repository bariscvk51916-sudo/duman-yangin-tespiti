```python
from google.colab import files
uploaded = files.upload()  # Roboflow ZIP dosyasını seç
zip_name = list(uploaded.keys())[0]
print('Yüklenen dosya:', zip_name)
```

```python
!unzip -q "{zip_name}" -d dataset
!find dataset -name "data.yaml" -o -name "dataset.yaml"
```

```python
from ultralytics import YOLO

# data.yaml yolunu bir üst hücredeki find çıktısına göre düzelt
DATA_YAML = "dataset/data.yaml"  # gerekirse güncelle

model = YOLO("yolo11n.pt")  # küçük ve hızlı model; yolo11s.pt de kullanılabilir
results = model.train(
    data=DATA_YAML,
    epochs=50,
    imgsz=640,
    batch=16,
    patience=10,
    project="runs",
    name="fire_smoke",
)
```

```python
from IPython.display import Image, display
display(Image(filename="runs/fire_smoke/results.png"))
display(Image(filename="runs/fire_smoke/confusion_matrix.png"))
```

```python
metrics = model.val()
print("mAP50:", metrics.box.map50)
print("mAP50-95:", metrics.box.map)
print("Precision:", metrics.box.mp)
print("Recall:", metrics.box.mr)
```
